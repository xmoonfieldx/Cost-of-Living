# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
dataset1 = pd.read_csv('cost-of-living.csv')
dataset = pd.read_csv('Cost_of_Living_Index_2022.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# Take care of the missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 2:])
X[:,2:] = imputer.transform(X[:, 2:])
```

# Splitting the dataset into the Training test and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler 
sc = MinMaxScaler()
X_train[:, 1:] = sc.fit_transform(X_train[:, 1:])
X_test[:, 1:] = sc.fit_transform(X_test[:, 1:])
```
