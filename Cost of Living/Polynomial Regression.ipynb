# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
dataset1 = pd.read_csv('cost-of-living.csv')
dataset = pd.read_csv('Cost_of_Living_Index_2022.csv')
X = dataset.iloc[1:, 2:-1].values
y = dataset.iloc[1:, -1].values
```

# Training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X[:,0], y, color = 'red')
plt.plot(X[:,0], lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X[:,0], y, color = 'red')
plt.plot(X[:,0], lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
X_grid = np.arange(min(X[:,0]), max(X[:,0]), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X[:,0], y, color = 'red')
plt.plot(X_grid[:,0], lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
